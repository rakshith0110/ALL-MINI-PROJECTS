```python
#Change value in rainfall column
df['rainfall'] = df['rainfall'].replace(['no'],'0')
df['rainfall'] = df['rainfall'].replace(['yes'],'1')
df['rainfall'] = pd.to_numeric(df['rainfall'])
```

```python
df.columns = df.columns.str.strip()
```

```python
sns.countplot(x=df['rainfall'])
print(df.rainfall.value_counts())
```

```python
import scipy.stats as stats
z = np.abs(stats.zscore(df))
dfo = df[(z<3).all(axis = 1)] 
dfo.shape
```

```python
from imblearn.over_sampling import SMOTE
X = dfo.drop('rainfall', axis=1)
y = dfo['rainfall']
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)
df_resampled = pd.concat([X_resampled, pd.Series(y_resampled, name='rainfall')], axis=1)
```

```python
sns.countplot(x=df_resampled['rainfall'])
print(df_resampled.rainfall.value_counts())
```

```python
X = dfo.drop('rainfall', 'day', axis=1)
y = dfo['rainfall']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2,random_state=0)
```

```python
from sklearn.metrics import accuracy_score,f1_score,precision_score,recall_score,confusion_matrix
```

```python
def model_fit(model):
    model.fit(X_train,y_train)
    y_pred = model.predict(X_test)
    model_acc =round(accuracy_score(y_pred,y_test)*100,2)
    print(f"Acurracy Score : {model_acc}%")
    print("")
    sns.heatmap(confusion_matrix(y_pred,y_test),annot=True,square=True,cmap='Blues')
    plt.xlabel("actual label")
    plt.ylabel("predicted label")
    plt.title(f"CONFUSION MATRIX")
    print("")
```

```python
#Logistic Regression(LR)
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(max_iter=1000, random_state=0)
model_fit(lr)
```

```python
#Support Vector Machine(SVC)
from sklearn.svm import SVC
svm = SVC(kernel='rbf', random_state=0)
model_fit(svm)
```

```python
#KNN
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
model_fit(knn)
```

```python
#XGBoost
from xgboost import XGBClassifier
xgb = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=0)
model_fit(xgb)
```

```python
#AdaBoost
from sklearn.ensemble import AdaBoostClassifier
ab = AdaBoostClassifier(random_state=0)
model_fit(ab)
```

```python
#RandomForest
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(random_state=0)
model_fit(rf)
```
